```python
df = pd.read_csv('/Users/teresazhang/Documents/NTU/Y3S1/BC3409/Assignment/Elearning/Groceries_dataset 2 (Association Rules)(1).csv')
```

```python
df = [a[1]['itemDescription'].tolist() for a in list(df.groupby(['Member_number','Date']))]
df
```

```python
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()
te_data = te.fit(df).transform(df)
df = pd.DataFrame(te_data,columns=te.columns_)
df
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
items = apriori(df, min_support = 0.01, use_colnames = True)
items
```

```python
rules = association_rules(items, metric ="confidence", min_threshold = 0.01)
rules
```

Support and confidence define a strong association rule. However, these measures might be biased. For the rule A → B, the confidence tends to be high when the support of B is high. Therefore, it is sometimes improper to interpret the rule as “A implies B” or use confidence as reliability measure.

Lift measures the correlation/dependence between item sets. Lift can an alternative measure to discover interesting association rules when the data is big.

If lift=1, A and B are independent (according to probability theory, support(A U B) = support(A) x support(B) given A and B are independent).
If lift > 1, A and B are positively correlated. Namely, the 2 item sets lift the likelihood of each another. In the grocery context, A and B are frequently bought together.
If lift < 1, A and B are negatively correlated. In the grocery context, the customers tend NOT to buy A and B together.
